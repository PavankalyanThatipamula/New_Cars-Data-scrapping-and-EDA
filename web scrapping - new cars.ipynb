```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
cars= ['tata-nexon',
 'maruti-suzuki-baleno',
 'tata-punch',
 'hyundai-grand-i10-nios',
 'hyundai-aura',
 'maruti-suzuki-wagon-r',
 'hyundai-new-i20',
 'maruti-suzuki-swift',
 'tata-tigor',
 'tata-altroz',
 'tata-tiago',
 'hyundai-venue',
 'hyundai-creta',
 'maruti-suzuki-celerio',
 'hyundai-alcazar',
 'hyundai-i20-n-line',
 'maruti-suzuki-dzire',
 'maruti-suzuki-s-presso',
 'nissan-magnite',
 'maruti-suzuki-brezza',
 'maruti-suzuki-brezza',
 'skoda-kodiaq',
 'honda-amaze',
 'honda-city',
 'maruti-suzuki-ignis',
 'maruti-suzuki-xl6',
 'maruti-suzuki-ertiga',
 'maruti-suzuki-ciaz',
 'tata-nexon-ev-prime',
 'tata-tiago-nrg',
 'maruti-suzuki-new-grand-vitara',
 'tata-nexon-ev-max',
 'maruti-suzuki-alto-k10',
 'mahindra-marazzo',
 'mahindra-bolero',
 'mahindra-scorpio-n',
 'mahindra-xuv700',
 'mahindra-xuv300',
 'toyota-urban-cruiser-hyryder',
 'toyota-glanza',
 'hyundai-tucson',
 'tata-tiago-ev',
 'tata-tigor-ev',
 'mahindra-xuv400',
 'mahindra-scorpio-classic',
 'maruti-suzuki-jimny',
 'hyundai-exter',
 'tata-harrier',
 'kia-carens',
 'mg-hector',
 'mg-gloster',
 'hyundai-verna',
 'kia-seltos',
 'mg-astor',
 'mg-hector-plus',
 'mg-zs-ev',
 'skoda-kushaq',
 'kia-sonet',
 'toyota-innova-hycross',
 'skoda-slavia',
 'volkswagen-virtus']
```

```python
variants =[]
price =[]
brand =[]
seater =[]
model =[]
fuel_type =[]
engine =[]
transmission =[]

for i in cars:
    url = f"https://ackodrive.com/cars/{i}/"
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    
    
    count = 0
    for j in soup.find_all("h4",class_="styles__VariantTitle-sc-ff1d2337-10 jrtfkd"):
        a = j.text
        count = count+1
        if a:
            variants.append(a)
        else:
            variants.append(np.nan)
    
    for k in range(count):
        a = soup.find("h1",class_="styles__MakeModelName-sc-3b7b7e29-1 iAQYbS")
        if a:
            brand.append(a.text)
        else:
            brand.append(np.nan)
            
    b = []
    for i in soup.find_all("div",class_="styles__Specs-sc-395b97e7-4 kjgGbG"):
      
        b.append(i.text.split(" ")[1])
    #print(b)
    for k in range(count):
        if b:
            model.append(b[-1])
        else:
            model.append(np.nan)
            
            
    c = []
    for i in soup.find_all("div",class_="styles__Specs-sc-395b97e7-4 kjgGbG"):
      
        c.append(i.text.split(" ")[1])
    #print(b)
    for k in range(count):
        if c:
            engine.append(c[1])
        else:
            engine.append(np.nan)    
            
    d = []
    for i in soup.find_all("div",class_="styles__Specs-sc-395b97e7-4 kjgGbG"):
      
        d.append(i.text.split(" ")[1])
    #print(b)
    for k in range(count):
        if d:
            seater.append(d[-2])
        else:
            seater.append(np.nan)      
 
        
    for j in soup.find_all("div",class_="styles__VariantInfo-sc-ff1d2337-12 ftPSTe"):
        #print(i.text)
        regex = re.findall("Petrol|Diesel|CNG|Electric",j.text)
        if regex:
            fuel_type.append(regex[0])
        else:
            fuel_type.append(np.nan)    
    for j in soup.find_all("div",class_="styles__VariantInfo-sc-ff1d2337-12 ftPSTe"):
       # print(i.text)
        regex = re.findall("(Automatic|Manual)",j.text)
        #print(regex[0])
        if regex:
            transmission.append(regex[0])
        else:
            transmission.append(np.nan) 
   
    for j in soup.find_all("div",class_="styles__Price-sc-ff1d2337-18 CHHgf"):
        #print(i.text)
        regex = re.findall("₹(\d+.?\d+)",j.text)
       # print(regex[0])
        if regex:
            price.append(regex[0])
        else:
            price.append(np.nan)        
```
